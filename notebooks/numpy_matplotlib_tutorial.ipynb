
# Learning NumPy and Matplotlib

This notebook will guide you through the basics of **NumPy arrays**, how to perform **masking**, do some **math operations**, 
and finally create an **interesting plot with Matplotlib**.

We will:
1. Create and manipulate NumPy arrays
2. Perform masking and filtering
3. Apply math operations
4. Define a function and an object (class)
5. Visualize results with Matplotlib


```python

import numpy as np

# Create a simple numpy array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Create a 2D array
mat = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", mat)

# Useful array properties
print("Shape:", mat.shape)
print("Data type:", mat.dtype)
print("Size:", mat.size)
```

```python

# Masking: select only values greater than 2
mask = arr > 2
print("Mask:", mask)

filtered = arr[mask]
print("Filtered values:", filtered)
```

```python

# Vectorized math
squared = arr ** 2
print("Squared:", squared)

# Apply sine function
angles = np.linspace(0, 2*np.pi, 10)
sin_vals = np.sin(angles)
print("Sine values:", sin_vals)
```

```python

# Define a mathematical function
def my_function(x):
    return np.sin(x) * np.exp(-0.1*x)

# Define a class that uses NumPy arrays
class Signal:
    def __init__(self, func, x):
        self.x = x
        self.y = func(x)
    
    def amplify(self, factor):
        self.y *= factor

# Create a signal
x = np.linspace(0, 10, 200)
signal = Signal(my_function, x)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(signal.x, signal.y, label="Original Signal")

# Amplify and plot again
signal.amplify(2)
plt.plot(signal.x, signal.y, label="Amplified Signal", linestyle="--")

plt.title("Signal Visualization with NumPy and Matplotlib")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
