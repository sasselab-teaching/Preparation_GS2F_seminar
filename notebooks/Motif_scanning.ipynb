```python
from pyjaspar import jaspardb
import numpy as np
import pandas as pd
# Download Jaspar motif database
jdb = jaspardb(release='JASPAR2024')
# fetch the human JUN motif (API-compatible across pyjaspar versions)
if hasattr(jdb, 'fetch_motif_by_id'):
    jun_motif = jdb.fetch_motif_by_id('MA0099.2')
elif hasattr(jdb, 'fetch_motifs_by_id'):
    motif_result = jdb.fetch_motifs_by_id('MA0099.2')
    jun_motif = motif_result[0] if isinstance(motif_result, list) else motif_result
else:
    motif_result = jdb.fetch_motifs_by_name('JUN')
    motifs = motif_result if isinstance(motif_result, list) else [motif_result]
    jun_motif = next((m for m in motifs if str(getattr(m, 'matrix_id', '')).startswith('MA0099.2')), motifs[0])
# print the counts matrix
print(jun_motif.counts)
# create pfm from counts (convert to dataframe)
counts_ = pd.DataFrame(jun_motif.counts)
print(counts_)
# normalize: divide each row by its sum so A+C+G+T=1 for each position
ppm = counts_.values / counts_.values.sum(axis=1, keepdims=True)
pwm = np.log2(ppm / 0.25)  # assuming uniform background
pwm[pwm < 0] = 0  # set negative values to zero for PWM
ppm = pd.DataFrame(ppm, columns=counts_.columns)
pwm = pd.DataFrame(pwm, columns=counts_.columns)

import matplotlib.pyplot as plt
import logomaker as lm 
# plot the jun motif as a sequence logo
fig, ax = plt.subplots(figsize=(len(pwm.index), 2))
lm.Logo(pwm, ax =ax)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
# free cuda memory
import torch
if torch.cuda.is_available():
    torch.cuda.empty_cache()
```

```python
# Let's scan the genome with this pwm
import pyfaidx
# load the reference genome
genome = pyfaidx.Fasta('../data/GRCh38/GRCh38.fa')

import torch
from torch.nn.functional import conv1d

def one_hot_encode_numpy(seq, dtype = np.float32):
    onehot = np.zeros((len(seq), 4), dtype=dtype)
    onehot[: len(seq)] = np.array(list(seq.upper()))[:, None] == np.array(['A', 'C', 'G', 'T'])
    onehot = onehot.T  # transpose to shape (4, seq_length)
    return onehot

# define device for torch (use GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# convert the pwm to a numpy array and reshape to (4, motif_length, 1) for conv1d
pwm_array = pwm.values.T[None, :, :]  # shape (1, 4, motif_length)
motif_length = pwm_array.shape[2]
pwm_tensor = torch.tensor(pwm_array, dtype=torch.float32).to(device)  # shape (1, 4, motif_length)

#iterate over the chromosomes in the genome and scan for the motif
pwm_scans = {}
for chrom in genome.keys():
    print(f'Scanning {chrom} for JUN motif..., pwm tensor shape: {pwm_tensor.shape}')
    seq = genome[chrom][:].seq
    print(f'Sequence length: {len(seq)}')
    # scan the sequence with the pwm with pytorch conv1d
    # convert the sequence to one-hot encoding
    onehot_seq = one_hot_encode_numpy(seq)
    print(f'One-hot encoded sequence shape: {onehot_seq.shape}')
    # perform convolution in chunks to avoid memory issues (scan 1 million bases at a time)
    chunk_size = 100000
    scores = []
    for i in range(0, onehot_seq.shape[-1], chunk_size):
        if i == 0:
            start = 0        
        else:
            start = i-motif_length//2 # include overlap for motif scanning
        chunk = onehot_seq[..., start:i+chunk_size]
        #print(chunk.shape)
        if i == 0:
            # pad the beginning of the first chunk with zeros to account for the motif length
            pad_width = motif_length//2
            chunk = np.pad(chunk, ((0, 0), (pad_width, pad_width)), mode='constant')
        else:
            # pad the end of the chunk with zeros to account for the motif length
            pad_width = motif_length//2
            chunk = np.pad(chunk, ((0, 0), (0, pad_width)), mode='constant')
        #print(start)
        #print(chunk.shape)
        onehot_seq_tensor = torch.tensor(chunk, dtype=torch.float32).unsqueeze(0).to(device)  # shape (1, 4, seq_length)
        chunk_scores = conv1d(onehot_seq_tensor, pwm_tensor).squeeze().detach().cpu().numpy()
        scores.append(chunk_scores)
    scores = np.concatenate(scores)
    print(f'Finished scanning {chrom} for JUN motif. Lenght of scores: {len(scores)}')
    pwm_scans[chrom] = scores
    break # only scan the first chromosome for testing


# save the pwm scans as bigwig files
import pyBigWig
for chrom, scores in pwm_scans.items():
    print(f'Saving PWM scan for {chrom} as bigwig...')
    bw = pyBigWig.open(f'../data/JUN_ChIP_ENCODE/{chrom}_JUN_pwm_scan.bw', 'w')
    bw.addHeader([(chrom, len(genome[chrom]))])
    # add entries using fixed step/span format
    bw.addEntries(chrom, 0, values=scores.tolist(), span=1, step=1)
    bw.close()

# check where the scan has the highest scores, and save the positions around as bed files
bed_file = open('../data/JUN_ChIP_ENCODE/JUN_pwm_scan_peaks.bed', 'w')
for chrom, scores in pwm_scans.items():
    max_score = np.max(scores)
    max_pos = np.argwhere(scores == max_score).flatten()
    print(f'Found {len(max_pos)} positions with max score {max_score} in {chrom}')
    for i in range(len(max_pos)):
        start = max_pos[i] - 3
        end = max_pos[i] + 4
        bed_file.write(f'{chrom}\t{start}\t{end}\t{scores[max_pos[i]]}\n')
bed_file.close()

    
```
